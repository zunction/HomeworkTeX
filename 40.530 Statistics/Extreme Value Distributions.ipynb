```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import os

from scipy.stats import invgauss, norm, gamma

from itertools import permutations

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) 
```

#### Extreme Values of Exp(1) random variables

```python
n = 10000
N = 10000
Mn_sample = np.zeros((N,1))
for i in range(N):
    Mn = np.max(np.random.exponential(size = n))
    Mn_sample[i] = Mn

plt.hist(Mn_sample, cumulative=True, normed=True, bins = 80, label='Histogram of Mn samples')
x = np.arange(0,25,0.1)
G = lambda x : np.exp(x)
y = G(-G(-x + np.log(n))) 
plt.plot(x,y, label = 'Analytic distribution')
plt.legend(loc='best')
# plt.savefig("1.png", dpi=1000)
```

#### Uniform variables, $U(0,1)$

```python
n = 10000
N = 10000
Mn_sample = np.zeros((N,1))
for i in range(N):
    Mn = np.max(np.random.uniform(0, 1, size = n))
    Mn_sample[i] = Mn

plt.hist(Mn_sample, cumulative=True, normed=True, bins = 80, label='Histogram of Mn samples')
x = np.arange(0, 1, 0.1)
G = lambda x : np.exp(x)
y = G(n * (x - 1)) 
plt.plot(x,y, label = 'Analytic distribution')
plt.legend(loc='best')
```

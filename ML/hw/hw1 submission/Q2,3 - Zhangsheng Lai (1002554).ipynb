```python
def load_data(path, feature = 'raw'):
    """
    Loads data into pixel values from the list of path given. Returns 
    Input:
    - path (list): list of path to load the data from.
    - feature: either 'raw' or 'hist' for raw pixel values and 3D histogram respectively.
    """
    x_train=[]
    y_train=[]
    for c,i in enumerate(path):
        os.chdir(i)
        l = os.listdir()
        for i in l:
            if feature == 'raw':
                vf = convert2pixel_value(i)
            else:
                vf = convert2color_3Dhist(i)
            x_train.append(vf)
            y_train.append(c)
    
    x_train = np.concatenate([i[np.newaxis] for i in x_train])
    y_train = np.array(y_train)
    
    # comment below to remove the shuffling of the data
    arr = np.arange(x_train.shape[0])
    np.random.shuffle(arr)
    x_train = x_train[arr]
    y_train = y_train[arr]
    
    return x_train, y_train
```

```python
def add_bias(dataset):
    """
    Add a one to each sample for bias. Dataset must be of the 
    form rows: samples, columns: features
    """
    n, m = dataset.shape
    out = np.ones((n, m+1))
    out[:,:-1] = dataset
    return out    
```

#### Using features obtained from the raw pixels to do the logistic loss 

```python
def sigmoid(x):
    """
    Applies the sigmoid function on the given vector.
    Input(s):
    - x : numpy vector of values
    """
    return 1/(1+np.exp(-x))
```

```python
def initialize_params(size=3073, seed=123):
    """
    Initialize parameters W weights and b biases.
    Input(s):
    - size (int): size of the parameters
    - seed (int): seed for the random number generator
    """
    rng = np.random.RandomState(seed)
    
    return rng.normal(size=(size,))
```

```python
def log_loss(x_train, y_train, W):
    """
    Computes the loss value of the logistic loss.
    Input(s):
    - x_train, y_train: training data and labels. x_train takes 
    different forms depending on the features used and y_train
    is {-1,+1}.
    - W: value of the parameters.                    
    """
    z = y_train * np.dot(x_train, W)
    h = sigmoid(z)
    
    return -np.mean(np.log(h))   
```

```python
def log_grad(x_train, y_train, W):
    """
    Computes the gradient of the logistic loss function.
    Input(s):
    - x_train, y_train: training data and labels. x_train takes 
    different forms depending on the features used and y_train
    is {-1,+1}.
    - W: value of the parameters.                    
    """
    z = y_train * np.dot(x_train, W)
    h = sigmoid(z)
    n = x_train.shape[0]
    
    return 1/n * np.dot(x_train.T,(y_train * (h-1)))    
```

```python
def next_batch(x_train, y_train, batch_size=2):
    """
    Returns a batch of size batch_size for stochastic gradient descent.
    - x_train, y_train: training data and labels. x_train takes different 
    forms depending on the features used and y_train is {-1,+1}.
    - batch_size (int): size of each batch.                    
    """
    for i in np.arange(0, x_train.shape[0], batch_size):
        yield (x_train[i:i+batch_size],y_train[i:i+batch_size])
```

```python
def log_classifier(x, learnt_W):
    """
    Takes in the test set and learnt parameters and returns the 
    accuracy of the classifier on the test set.
    Inputs:
    - x: data for classification
    - learnt_W: learnt parameters 
    """
    return (sigmoid(np.dot(x, learnt_W)) >= .5) * 2 - 1
```

```python
def log_accuracy(x, y, learnt_W):
    """
    Returns the accuracy of the model with parameters learnt_W.
    Input(s):
    - x: data for classification
    - y: corresponding labels to the data x
    - learnt_W: learnt parameters 
    """
    output = log_classifier(x, learnt_W)
    return np.sum(np.absolute(y - output) == 0)/y.shape[0]
```

```python
def log_train(x_train, y_train, x_test, y_test, W, alpha=0.01, batch_size = 4, epoch = 100):
    """
    Trains the log loss model with given learning rate alpha, batch_size, epoch and initialized parameters W.
    Returns the history of the loss, train accuracy, test accuracy and the value of the parameters 
    at different epochs.
    Input(s):
    - x_train, y_train: train data and labels
    - x_test, y_test: test data and labels
    - W: parameters of the model
    - alpha: learning rate
    - batch_size: batch size for stochastic gradient descent
    - epoch: number of times the dataset is passed through the model.
    """
    loss_history = []
    train_acc_history = []
    test_acc_history = []
    W_history = []
    for e in np.arange(epoch):
        epoch_loss = []
        for (batchx, batchy) in next_batch(x_train, y_train):
            loss = log_loss(batchx, batchy, W)
            grad = log_grad(batchx, batchy, W)
            epoch_loss.append(loss)
            W += -alpha * grad
        loss_history.append(np.average(epoch_loss))
        train_acc = log_accuracy(x_train, y_train, W)
        test_acc = log_accuracy(x_test, y_test, W)
        train_acc_history.append(train_acc)
        test_acc_history.append(test_acc)
        W_temp = W.copy()
        W_history.append(W_temp)
           
    return loss_history, train_acc_history, test_acc_history, W_history    
```

```python
def display_plot(loss_history, train_acc_history, test_acc_history):
    """
    Plots out the loss function value for the different epochs and the accuracy 
    of the test set for the parameters learnt in the different epochs.
    Input(s):
    - loss_history: list containing the change in the loss value over the epochs
    - train_acc_history: list containing the accuracy value of the train set over the epochs
    - test_acc_history: list containing the accuracy value of the test set over the epochs
    """
    f, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,5))
    ax1.plot(loss_history, color='r')
    # ax1.plot(model_log.history['val_acc'])
    ax1.set_title('Loss (Lower Better)')
    ax1.set(xlabel='Epoch', ylabel='Loss')
    ax1.legend(['train', 'validation'], loc='upper right')

    ax2.plot(train_acc_history)
    ax2.plot(test_acc_history)
    ax2.set_title('Accuracy (Higher Better)')
    ax2.set(xlabel='Epoch', ylabel='Accuracy')
    ax2.legend(['train', 'test'], loc='center right')
```

Here `W1` refers to the initialized parameters that will be used to learn the model from using the raw pixel values as the training data.

```python
W1 = initialize_params(seed=1)
loss_history1, train_acc_history1, test_acc_history1, learnt_W_history1 = log_train(x1_train, y1_train, 
                                                                                    x1_test, y1_test, 
                                                                                    W1, epoch=100)
```

```python
display_plot(loss_history1, train_acc_history1, test_acc_history1)
```

```python
def para_diff(learnt_W_history):
    """
    Plots the change in the value of the parameter learnt and returns the epoch 
    with the smallest change.
    Input(s):
    - learnt_W_history: history of the different value of the parameter 
    at different epoch.
    """
    diff = []
    for i in range(0,len(learnt_W_history)-1):
        d = np.linalg.norm(learnt_W_history[i]-learnt_W_history[i+1])/np.linalg.norm(learnt_W_history[i])
        diff.append(d)

    plt.plot(diff)
    plt.title('Change in parameter')
    plt.xlabel('Epoch')
    plt.ylabel('Difference')
    print ('Epoch with minimum change in parameter: {} '.format(np.argmin(diff)))
```

```python
para_diff(learnt_W_history1)
```

Using logistic loss and raw pixel features, we see from the plot above, which is generated from $\frac{||\theta_{k+1}-\theta_{k}||}{||\theta_{k}||}$ we see that the change in the parameters become less significant when the epoch reaches the range of 70s, with 50% accuracy, no better than random guessing. Thus we can stop the iteration in that range as it has arrived at a local optimum. This does not necessary imply that that perforance on the test set, measured by the test accuracy has high accuracy as the local optimum might not be the global or a good local optimum.

#### Discussion:
From the experiments above, we see that the test accuracy is not a good indicator of when to stop the experiment as due to the constrain of our model, we are unable to arrive at high test accuracy. Using the loss function is also not necessarily a good indicator for stopping as a small loss function value might cause the model the overfit. Thus we look to how much the parameter change to see if the learning has reached a local optimal. If the change in the parameter value is small, it means that the gradient of the point in the parameter space is close to a stationary point and thus it is not meaningful to continuing running the iterations. However, the parameters that we arrive at might not be the desirable ones as there might be other optimals that may give better test accuracy. The way to allow the algorithm to reach other optimals which might be better is to start with different initial parameters. For the sake of experiment, I have used the same initial parameters, but using different random initial parameters might possible allow the learning to arrive at better optimals.

#### **Q3.** Hinge loss with stochastic gradient descent to perform binary classification

As in earlier, we will use `1` and `2` labels to reference using the raw pixel values and the features from the histogram respectively.

Here we shall use `theta` to represent the parameters.

```python
def hinge_loss(x_train, y_train, theta):
    """
    Returns the hinge loss function values
    Input(s):
    - x_train, y_train: train data and labels
    - theta: parameters for the model
    """
    z = y_train * np.dot(x_train, theta)
    return np.average(np.maximum(1-z,0))
```

```python
def hinge_grad(x_train, y_train, theta):
    """
    Returns the gradient of the hinge loss function at theta
    Input(s):
    - x_train, y_train: train data and labels
    - theta: parameters for the model
    """
    z = y_train * np.dot(x_train, theta)
    grad = -(y_train * x_train.T).T
    grad[z>1]=0
    return np.average(grad, axis=0)
```

```python
def hinge_classifier(x, learnt_theta):
    """
    Performs classification using hinge loss
    Input(s):
    - x: data for classification
    - learnt_theta: learnt parameters 
    """
    return np.sign(np.dot(x,learnt_theta))
```

```python
def hinge_accuracy(x, y, learnt_theta):
    """
    Returns accuracy of the hinge loss
    Input(s):
    - x: data for classification
    - y: label of the data
    - learnt_W: learnt parameters 
    """
    output = hinge_classifier(x, learnt_theta)
    return np.sum(np.absolute(y - output) == 0)/y.shape[0]   
```

```python
def hinge_train(x_train, y_train, x_test, y_test, theta, alpha=0.01, batch_size = 1, epoch = 100):
    """
    Trains the hinge loss model with given learning rate alpha, batch_size, epoch and initialized parameters W.
    Returns the history of the loss, train accuracy, test accuracy and the value of the parameters 
    at different epochs.
    Input(s):
    - x_train, y_train: train data and labels
    - x_test, y_test: test data and labels
    - W: parameters of the model
    - alpha: learning rate
    - batch_size: batch size for stochastic gradient descent
    - epoch: number of times the dataset is passed through the model.
    """
    loss_history = []
    train_acc_history = []
    test_acc_history = []
    theta_history =[]
    for e in np.arange(epoch):
        epoch_loss = []
        for (batchx, batchy) in next_batch(x_train, y_train):
            loss = hinge_loss(batchx, batchy, theta)
            grad = hinge_grad(batchx, batchy, theta)
            epoch_loss.append(loss)
            theta += -alpha * grad
        loss_history.append(np.average(epoch_loss))
        train_acc = log_accuracy(x_train, y_train, theta)
        test_acc = log_accuracy(x_test, y_test, theta)
        train_acc_history.append(train_acc)
        test_acc_history.append(test_acc)
        theta_temp = theta.copy()
        theta_history.append(theta_temp)
    
    return loss_history, train_acc_history, test_acc_history, theta_history
```

```python
theta1 = initialize_params(size = x1_train.shape[1])
loss_history1, train_acc_history1, test_acc_history1, learnt_theta_history1 = hinge_train(x1_train, y1_train, x1_test, y1_test, theta1, epoch = 100)
```

```python
display_plot(loss_history1, train_acc_history1, test_acc_history1)
```

```python
para_diff(learnt_theta_history1)
```

Using hinge loss and the raw pixel values as the features, we see from the plots above, that the learning arrives at a local optimum in the range of 70s, explained by the plateauing of the test accuracy at 55% and also the change in the values of the parameter learnt.
